```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("data/raw/Dataset_SCO_KVM_MONS_GRC_IZD_DMA_MTK_20250805.csv")
df2 = pd.read_csv("data/raw/Etagi_secondary_classified_dataset_20250805.csv")
df3 = pd.read_csv("data/raw/Etagi_secondary_dataset_20250805.csv")
```

```python
print(f"df1: {df1.shape[0]} x {df1.shape[1]}")
print(f"df2: {df2.shape[0]} x {df2.shape[1]}")
print(f"df3: {df3.shape[0]} x {df3.shape[1]}")
```

```python
pd.set_option('display.max_columns', None)
```

```python
df1_no_dups = df1.drop_duplicates()
df2_no_dups = df2.drop_duplicates()
df3_no_dups = df3.drop_duplicates()
```

```python
df1_no_empty = (
    df1_no_dups.replace('', pd.NA)
       .dropna(how='all')
       .dropna(axis=1, how='all')
)

df2_no_empty = (
    df2_no_dups.replace('', pd.NA)
       .dropna(how='all')
       .dropna(axis=1, how='all')
)

df3_no_empty = (
    df3_no_dups.replace('', pd.NA)
       .dropna(how='all')
       .dropna(axis=1, how='all')
)
```

```python
df1_del_columns = df1_no_empty.drop(columns=['ID помещения'])
df2_del_columns = df2_no_empty.drop(columns=['object_id', 'url', 'parent_id', 'duplicates_url', 'city_id'])
df3_del_columns = df3_no_empty.drop(columns=['source_city_id', 'city_id'])
```
