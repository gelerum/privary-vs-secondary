```python
df = pd.read_parquet('msk_united_geo_market_deals.parquet')
```

```python
import folium


# Случайная выборка 50000 точек
sample_df = df.sample(n=50000, random_state=42)

# Дальше используем, например, Plotly или Folium

# Проверяем названия колонок
print(sample_df.columns.tolist())

# Создаем карту с центром в средних координатах
center_lat = sample_df['latitude'].mean()  # замените на ваше название колонки
center_lon = sample_df['longitude'].mean()  # замените на ваше название колонки

m = folium.Map(location=[center_lat, center_lon], zoom_start=10)

# Добавляем точки
for idx, row in sample_df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        popup=f"Point {idx}<br>Lat: {row['latitude']}<br>Lon: {row['longitude']}",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(m)

# Сохраняем карту в HTML файл
m.save('map.html')
print("Карта сохранена как map.html")
```
